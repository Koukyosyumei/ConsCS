```python
import sys
import os

parent_dir = os.path.abspath('../')  
sys.path.append(parent_dir)

from utils import *
```

```python
def run_and_log_and_count_poly(target_folder, file_path, log_path, log_path_with_contributions):  
    if not file_path.endswith('.r1cs'):
        return dict()
    
    if file_path in ['Main@main@circomlib.r1cs', 'Sha256_2@sha256_2@circomlib.r1cs', 'Sha256compression@sha256compression@circomlib.r1cs']:
        # extremely large, times out without finishing reading
        return dict()
    
    
    
    file_name = os.path.basename(file_path)
    base_name = file_name.split('.')[0]
    full_path = os.path.join(target_folder, file_path)
    target_file_path = full_path
    print(target_file_path)
    
    cur_d = dict()
    
    clear_data_structures()
    r1cs_circuit = parse_entire_circuit(target_file_path, print_all_circuits = 0)
    for cons in r1cs_circuit.all_constraints:
        cur_len = len(cons.terms) - 1
        cur_d[cur_len] = cur_d.get(cur_len, 0) + 1

    return cur_d

```

```python
all_d = dict()
all_d_counts = dict()
log_path = ''
log_path_with_contributions = ''
for type_folder in ["utils", "core"]:
    target_folder = os.path.join("../benchmarks/", type_folder)
    for file_path in os.listdir(target_folder):
        cur_d = run_and_log_and_count_poly(target_folder, file_path, log_path, log_path_with_contributions)
        assert file_path not in all_d
        all_d[file_path] = cur_d
        for k in cur_d:
            all_d_counts[k] = all_d_counts.get(k, 0) + cur_d[k]
                   
```

```python
all_d_counts
```

```python
for max_depth in range(10):
    all_counts = 0
    short_counts = 0
    for k in all_d_counts:
        all_counts += all_d_counts[k]
        if k <= max_depth:
            short_counts += all_d_counts[k]
    print(f"depth {max_depth}", short_counts, all_counts, short_counts/all_counts * 100)
```
